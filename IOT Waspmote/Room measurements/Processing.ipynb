```python
ind_bat = []
ind_co2 = []
ind_temp = []
ind_hum = []
ind_pres = []
```

```python
for index, row in data_new.iterrows():
    if row[0] == 'BAT':
        ind_bat.append(index)
    if row[0] == 'CO2':
        ind_co2.append(index)
    if row[0] == 'TC':
        ind_temp.append(index)
    if row[0] == 'HUM':
        ind_hum.append(index)
    if row[0] == 'PRES':
        ind_pres.append(index)
```

```python
data_bat = data_new.loc[ind_bat]
data_co2 = data_new.loc[ind_co2]
data_temp = data_new.loc[ind_temp]
data_hum = data_new.loc[ind_hum]
data_pres = data_new.loc[ind_pres]
```

```python
data_bat = data_bat.drop(columns='BAT')
data_co2 = data_co2.drop(columns='BAT')
data_temp = data_temp.drop(columns='BAT')
data_hum = data_hum.drop(columns='BAT')
data_pres = data_pres.drop(columns='BAT')
```

```python
data_bat = data_bat.rename(columns={"16": "BAT", "2019-11-27 14:33:07": "date"})
data_co2 = data_co2.rename(columns={"16": "CO2", "2019-11-27 14:33:07": "date"})
data_temp = data_temp.rename(columns={"16": "TEMP", "2019-11-27 14:33:07": "date"})
data_hum = data_hum.rename(columns={"16": "HUM", "2019-11-27 14:33:07": "date"})
data_pres = data_pres.rename(columns={"16": "PRES", "2019-11-27 14:33:07": "date"})
```

```python
def index_rename(data_bat):
    return (data_bat.rename(index=lambda s: ((s + 1) // 5)))
```

```python
data_bat = index_rename(data_bat)
data_co2 = index_rename(data_co2)
data_temp = index_rename(data_temp)
data_hum = index_rename(data_hum)
data_pres = index_rename(data_pres)
```

```python
data_bat
```

```python
date = pd.DataFrame(data_bat.date)
bat = pd.DataFrame(data_bat.BAT)
co2 = pd.DataFrame(data_co2.drop(columns=date))
temp = pd.DataFrame(data_temp.TEMP)
hum = pd.DataFrame(data_hum.HUM)
pres = pd.DataFrame(data_pres.PRES)
```

```python
# data_finish = data_bat(columns='date')
data_finish = date.join(bat)
data_finish = data_finish.join(temp)
data_finish = data_finish.join(co2)
data_finish = data_finish.join(hum)
data_finish = data_finish.join(pres)
data_finish
```
